```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# vectorial function
logistica = np.frompyfunc(lambda b0, b1, x:
                            1 / (1 + math.exp(-(b0 + b1 * x))), # three parameters from logistic regression formula
                            3, 1)  
```

```python
plt.figure(figsize=(10,5))
plt.scatter(np.arange(-5, 5, 0.1), logistica(0,1, np.arange(-5, 5, 0.1)), color = "green") #we indicate logistica(b0, b1, x)

plt.scatter(np.arange(-5, 5, 0.1), logistica(0,2, np.arange(-5, 5, 0.1)), color = "gold") #the same but with different b1 ('slope' = 'pendiente')

plt.scatter(np.arange(-5, 5, 0.1), logistica(0,3, np.arange(-5, 5, 0.1)), color = "red") 

plt.title("Función logística Estándar - Diferentes 'pendientes'", fontsize= 14.0)
plt.ylabel("Probabilidad", fontsize = 13.0)
plt.show()
```

# Clasification with logistic regression

We are going to classify our data in normal or o not normal (tachycardia)

```python
heart_rate_nor = [65, 70, 80, 80, 80,
                   90, 95, 100, 105, 110]

heart_rate_tac = [105, 110, 110, 120, 120,
                        130, 140, 180, 185, 190]
```

```python
# Drawing a logistic function
# Y = b0 + b1x
# y = - 46.68057196 + 0.42460226

plt.scatter(np.arange(60,200,0.1),                      # from 60 to 200 is the range of the cardiac frequency
            logistica(-46.68057196, 0.42460226,        # our logistc reg
            np.arange(60,200,0.1)))

```

```python
# Graphic with heart rate in normal and in tachycardia
plt.scatter(heart_rate_nor, [0]*10, marker = "o", c= "green", s=250, label = "Normal")
```

```python

plt.scatter(heart_rate_tac, [0]*10, marker = "o", c= "blue", s=250, label = "Tachycardia")
```

# Everything together

```python

# Graphic with heart rate in normal and in tachycardia
plt.scatter(heart_rate_nor, [0]*10, marker = "o", c= "green", s=250, label = "Normal")

plt.scatter(heart_rate_tac, [1]*10, marker = "o", c= "blue", s=250, label = "Tachycardia")
# probability for 3 individuals
plt.scatter(np.arange(60,200,0.1),                      # from 60 to 200 is the range of the cardiac frequency
            logistica(-46.68057196, 0.42460226,        # our logistc reg
            np.arange(60,200,0.1)))


individual = [80, 110, 180]
probability = logistica(-46.68057196, 0.42460226, individual)

plt.scatter(individual, probability, marker = "*", c ="darkorange", s = 500)

plt.text(individual[0]+7, 0.05, "%0.2f" % probability[0], size = 12, color ="black")

plt.text(individual[1]+7, 0.485, "%0.2f" % probability[1], size = 12, color ="black")

plt.text(individual[2]+7, 0.90, "%0.2f" % probability[2], size = 12, color ="black")

plt.text(0,1, "Tachycardia", size= 12, color = "black")
plt.text(0,0, "Normal", size= 12, color = "black")

plt.ylabel("Probability of Tachycardia", fontsize = 13)
plt.xlabel("Heart rate (beat per minute)", fontsize = 13)

plt.legend(bbox_to_anchor=(1, 0.2))
plt.show()

```
